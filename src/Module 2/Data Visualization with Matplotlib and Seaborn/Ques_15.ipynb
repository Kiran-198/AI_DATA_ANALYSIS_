```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load iris dataset
iris = sns.load_dataset('iris')

# ✅ Select only numeric columns for correlation
corr = iris.select_dtypes(include='number').corr()

# Create heatmap with advanced styling
sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'},
    square=True
)

plt.title('Correlation Matrix Heatmap - Iris Dataset', fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
