```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Set custom font properties
title_font = {'fontsize': 16, 'fontweight': 'bold'}
label_font = {'fontsize': 12}

# Create bar plot
plt.bar(products, sales, color='teal')

# Add grid
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Add labels and title with custom font
plt.xlabel('Products', fontdict=label_font)
plt.ylabel('Sales (in thousands)', fontdict=label_font)
plt.title('Customized Product Sales Plot', fontdict=title_font)

# Show plot
plt.show()
```
