```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Create line plot with custom styles
plt.plot(x, y, color='purple', linestyle='--', marker='o', markerfacecolor='orange', markersize=8, linewidth=2)

# Add grid with customization
plt.grid(True, linestyle=':', linewidth=0.8, alpha=0.7)

# Add title and labels with font customization
plt.title('Customized Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.show()
```
