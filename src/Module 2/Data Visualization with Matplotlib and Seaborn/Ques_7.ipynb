```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]  # size variable

# Create DataFrame
data = pd.DataFrame({
    'Height (cm)': heights,
    'Weight (kg)': weights,
    'Size': sizes
})

# Create enhanced scatter plot
sns.scatterplot(
    x='Height (cm)',
    y='Weight (kg)',
    size='Size',
    data=data,
    sizes=(50, 300),       # scale of point sizes
    color='steelblue',
    legend=False           # hide legend for size
)

# Add title
plt.title('Height vs Weight with Size-based Points (Seaborn)')

# Show plot
plt.show()
```
