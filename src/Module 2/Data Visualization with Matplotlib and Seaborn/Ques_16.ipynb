```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y_line = [2, 3, 5, 7, 11]
y_bar = [5, 7, 3, 8, 6]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [5, 4, 6, 5, 7]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (top-left)
axs[0, 0].plot(x, y_line, color='blue', marker='o')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Bar plot (top-right)
axs[0, 1].bar(x, y_bar, color='orange')
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Scatter plot (bottom-left)
axs[1, 0].scatter(x_scatter, y_scatter, color='green', s=80)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Hide the bottom-right subplot (empty)
axs[1, 1].axis('off')

# Adjust layout for neatness
plt.tight_layout()

plt.show()
```
