```python
# Title: Load & Examine a Basic Dataset
# Description: Load an example dataset such as the Iris dataset and perform basic exploration.
# Load & Examine a Basic Dataset - Iris

# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert to a DataFrame for easier handling
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column
iris_df['target'] = iris.target

# Mapping target to actual class names
iris_df['species'] = iris_df['target'].map(dict(enumerate(iris.target_names)))

# Display the first 5 rows of the dataset
print("First 5 rows of the Iris dataset:")
print(iris_df.head())

# Basic information
print("\nDataset Info:")
print(iris_df.info())

# Summary statistics
print("\nSummary Statistics:")
print(iris_df.describe())

# Count of each class
print("\nClass Distribution:")
print(iris_df['species'].value_counts())
```
