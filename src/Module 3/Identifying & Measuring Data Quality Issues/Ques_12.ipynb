```python
# Data Profiling Techniques Examples

# 1. Descriptive Statistics:
# Task 1: Calculate the mean, median, and mode for sales figures in a retail dataset.
# Task 2: Analyze the average age, median, and mode in a customer demographic
# dataset.
# Task 3: Determine the mean, median, and mode of daily website visit counts.

import pandas as pd

# Task 1: Retail sales figures
sales_data = pd.DataFrame({'sales': [250, 300, 220, 400, 300, 275, 310, 300]})
print("Sales - Mean:", sales_data['sales'].mean())
print("Sales - Median:", sales_data['sales'].median())
print("Sales - Mode:", sales_data['sales'].mode()[0])

# Task 2: Customer age analysis
age_data = pd.DataFrame({'age': [25, 32, 28, 40, 28, 35, 28, 42]})
print("Age - Mean:", age_data['age'].mean())
print("Age - Median:", age_data['age'].median())
print("Age - Mode:", age_data['age'].mode()[0])

# Task 3: Website visit counts
visit_data = pd.DataFrame({'visits': [1200, 1100, 1300, 1250, 1100, 1400, 1100]})
print("Visits - Mean:", visit_data['visits'].mean())
print("Visits - Median:", visit_data['visits'].median())
print("Visits - Mode:", visit_data['visits'].mode()[0])





# 2. Distribution Analysis:
# Task 1: Visualize the distribution of product prices using histograms.
# Task 2: Create a distribution chart for exam scores in an educational dataset.
# Task 3: Analyze the distribution of order quantities for inventory management.


import seaborn as sns
import matplotlib.pyplot as plt

# Task 1: Product prices histogram
product_prices = pd.DataFrame({'price': [15, 20, 22, 20, 19, 25, 30, 22, 22]})
sns.histplot(product_prices['price'], kde=True)
plt.title("Product Price Distribution")
plt.show()

# Task 2: Exam scores distribution
exam_scores = pd.DataFrame({'score': [65, 70, 85, 90, 75, 80, 70, 68, 77]})
sns.histplot(exam_scores['score'], bins=5, kde=True)
plt.title("Exam Scores Distribution")
plt.show()

# Task 3: Order quantities distribution
order_quantities = pd.DataFrame({'quantity': [10, 12, 15, 12, 11, 20, 18, 12]})
sns.histplot(order_quantities['quantity'], kde=True)
plt.title("Order Quantities Distribution")
plt.show()




# 3. Data Type Validation:
# Task 1: Validate numeric fields in a dataset to ensure they contain only numbers.
# Task 2: Check for valid date formats in a transaction log.
# Task 3: Validate email formats in a customer contact dataset.
import re
from datetime import datetime

# Task 1: Numeric validation
df = pd.DataFrame({'amount': ['100', '250.5', 'abc', '300', '400.75']})
df['is_numeric'] = df['amount'].apply(lambda x: str(x).replace('.', '', 1).isdigit())
print("Numeric Validation:\n", df)

# Task 2: Date format validation
date_df = pd.DataFrame({'transaction_date': ['2022-01-01', '01-02-2022', '2022/03/01', 'March 5, 2022']})
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str)
        return True
    except:
        return False
date_df['valid_date'] = date_df['transaction_date'].apply(is_valid_date)
print("\nDate Format Validation:\n", date_df)

# Task 3: Email validation
email_df = pd.DataFrame({'email': ['test@example.com', 'invalid-email', 'user@domain.org', 'hello@', 'john.doe@gmail.com']})
def is_valid_email(email):
    return re.match(r"[^@]+@[^@]+\.[^@]+", email) is not None
email_df['valid_email'] = email_df['email'].apply(is_valid_email)
print("\nEmail Format Validation:\n", email_df)







```
