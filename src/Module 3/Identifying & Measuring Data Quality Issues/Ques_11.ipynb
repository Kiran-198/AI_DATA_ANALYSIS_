```python
# Common Data Errors Examples

# 1. Missing Data:
# Task 1: Review a dataset where some customer emails are missing. Identify how
# many records are incomplete.
# Task 2: Examine a sales dataset with missing transaction dates and determine the
# percentage of missing data.
# Task 3: Identify missing department information in an employee registry.

import pandas as pd

# Example dataset
df = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Email': ['a@example.com', None, 'c@example.com']
})

# Task 1: Count records with missing emails
missing_emails = df['Email'].isnull().sum()
print("Missing emails:", missing_emails)







# 2. Duplicate Data:
# Task 1: Analyze a customer dataset with duplicate entries and count the number of
# duplicates.
# Task 2: Review supplier data and identify any repeated supplier names.
# Task 3: Examine a product inventory list for duplicates in product IDs.


# Example dataset
df = pd.DataFrame({
    'CustomerID': [1, 2, 1],
    'Name': ['Alice', 'Bob', 'Alice']
})

# Task 1: Count duplicate rows
duplicate_rows = df.duplicated().sum()
print("Duplicate rows:", duplicate_rows)

# Task 2: Count duplicate names
duplicate_names = df.duplicated(subset='Name').sum()
print("Duplicate names:", duplicate_names)

# Task 3: Check for duplicate IDs
duplicate_ids = df.duplicated(subset='CustomerID').sum()
print("Duplicate IDs:", duplicate_ids)




# 3. Inconsistent Formatting:
# Task 1: Spot inconsistencies in date formats (e.g., DD/MM/YYYY vs. MM/DD/YYYY)
# in a dataset.
# Task 2: Identify phone numbers with varying formats in a contact list.
# Task 3: Review address data for discrepancies in state abbreviations (e.g., CA vs.
# Calif.).


# Task 1: Inconsistent date formats
df = pd.DataFrame({
    'Date': ['12/31/2024', '31/12/2024']
})

# Try parsing and catching errors
for date in df['Date']:
    try:
        parsed = pd.to_datetime(date, dayfirst=False)
        print(f"Parsed (US format): {parsed}")
    except:
        print(f"Inconsistent format: {date}")

# Task 2: Inconsistent phone formats
df = pd.DataFrame({
    'Phone': ['(123) 456-7890', '123-456-7890', '+1-123-456-7890']
})

# Simple print of unique formats
print("Unique phone formats:", df['Phone'].unique())

# Task 3: State abbreviation inconsistency
df = pd.DataFrame({
    'State': ['CA', 'Calif.', 'California']
})

print("Unique state formats:", df['State'].unique())



# 4. Data Drift:
# Task 1: Compare monthly revenues over six months to identify data drift.
# Task 2: Analyze user engagement metrics from a web application over different
# quarters.
# Task 3: Review a stock price dataset to detect any anomalies over a year.

# Task: Monthly Revenue Drift
df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [50000, 51000, 49800, 70000, 71000, 72000]
})

# Plotting drift (requires matplotlib)
import matplotlib.pyplot as plt

plt.plot(df['Month'], df['Revenue'], marker='o')
plt.title("Monthly Revenue Drift")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()


```
