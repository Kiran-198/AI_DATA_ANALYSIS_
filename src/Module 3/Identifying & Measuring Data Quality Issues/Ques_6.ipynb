```python
# Question: Evaluating Data Distribution
# Description: Analyze the distribution of a numeric column using histograms and descriptive statistics.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (example: Iris dataset)
df = sns.load_dataset('iris')  # Replace this with pd.read_csv('your_file.csv') if using a custom dataset

# Select the numeric column to analyze
numeric_column = 'sepal_length'  # Change this to any numeric column in your dataset

# Descriptive statistics
print(f"Descriptive Statistics for '{numeric_column}':")
print(df[numeric_column].describe())

# Plot the histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(df[numeric_column], kde=True, color='skyblue', bins=20)
plt.title(f'Distribution of {numeric_column}')
plt.xlabel(numeric_column)
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
